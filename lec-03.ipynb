```python
import numpy as np
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0   # first we will start the random value of m and b
    iterations = 1000  # no of steps
    
    n = len(x)
    learning_rate = 0.08  # randomly trial error
    
    for i in range(iterations):
        yp = m_curr * x + b_curr
        
        cost = (1/n) * sum([val ** 2 for val in (y-yp)])  # sum of square of every delta value
        
        # partial derivate
        md = -(2/n) * sum(x*(y-yp))
        bd = -(2/n) * sum(y-yp)
        
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        print(f"m {m_curr}, b {b_curr}, cost {cost} iteration {i}")
        
    pass
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
```

```python
gradient_descent(x, y)
```

## Visual Representation

```python
# we will use the Gradient descent to find the best fit line
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.01
    n = len(x)
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
#         print (m_curr,b_curr, i)
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
