```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/canada_per_capita_income.csv")
```

```python
df
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
new_df
```

```python
per_capita_income = df.drop('year', axis='columns')
```

```python
per_capita_income
```

```python
# Create linear regression object

reg = linear_model.LinearRegression()
reg.fit(new_df, per_capita_income)
```

```python
# predict canada's per capita income in year 2020
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2021]])
```
